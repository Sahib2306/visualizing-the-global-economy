# ðŸ“ˆ Tracking the Rise and Shifts of Global Economic Growth

A comparative view of GDP growth across major world economies, revealing how growth momentum  
changes over time and across countries.

This visualization uncovers:

- Growth volatility  
- Diverging economic cycles  
- Structural growth differences

```python
import plotly.express as px

# selecting some important countries 
selected_countries = ["United States","China","India","Germany","United Kingdom"]

filtered_df = df[df['country_name'].isin(selected_countries)]

fig = px.line(
    filtered_df,
    x='year',
    y="GDP Growth (% Annual)",
    color="country_name",
    title="GDP Growth Comparison Across Major Economies",

)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Year",
    yaxis_title="GDP Growth (%)",
    title_x=0.5
)

fig.show()
```

# ðŸ•¸ Multi-Indicator Economic Profile

Radar charts provide a compact way to compare countries across multiple  
macroeconomic dimensions simultaneously.

This visualization highlights structural differences in:

- Economic Growth  
- Inflation  
- Unemployment  
- Public Debt  
- Interest Rates  


```python
import pandas as pd
import plotly.express as px

selected_countries = ["United States", "China", "India", "Germany", "United Kingdom"]

radar_df = df[
    (df["country_name"].isin(selected_countries)) &
    (df["year"] == 2020)
].copy()

radar_df = radar_df.dropna(
    subset=[
        "GDP Growth (% Annual)",
        "Inflation (CPI %)",
        "Unemployment Rate (%)",
        "Public Debt (% of GDP)",
        "Interest Rate (Real, %)"
    ]
)

# Convert to long format (THIS FIXES EVERYTHING)
radar_long = radar_df.melt(
    id_vars=["country_name"],
    value_vars=[
        "GDP Growth (% Annual)",
        "Inflation (CPI %)",
        "Unemployment Rate (%)",
        "Public Debt (% of GDP)",
        "Interest Rate (Real, %)"
    ],
    var_name="Indicator",
    value_name="Value"
)

fig = px.line_polar(
    radar_long,
    r="Value",
    theta="Indicator",
    color="country_name",
    line_close=True,
    title="Comparative Economic Structure (2020)"
)

fig.update_layout(template="plotly_white", title_x=0.5)

fig.show()
```

# ðŸ† Leading Economies by GDP

This visualization ranks countries based on total economic output (GDP).

Bar charts provide a clear comparison of economic scale and highlight  
global leaders in production and income generation.


```python

year_filtered = df[df["year"] == 2020].copy()

year_filtered = year_filtered.dropna(subset=["GDP (Current USD)"])

top_gdp = year_filtered.sort_values(
    "GDP (Current USD)", ascending=False
).head(20)

fig = px.bar(
    top_gdp,
    x="country_name",
    y="GDP (Current USD)",
    title="Top 20 Economies by GDP (2020)",
)

fig.update_layout(
    template="plotly_white",
    title_x=0.5,
    xaxis_title="Country",
    yaxis_title="GDP (Current USD)"
)

fig.show()
```

# ðŸ› Fiscal Dynamics: Government Revenue vs Expenditure

Government revenue and expenditure as a percentage of GDP illustrate  
a country's fiscal structure and policy stance.

This visualization compares how governments balance income and spending.

```python
year_filtered = df[df["year"] == 2020].copy()

year_filtered = year_filtered.dropna(
    subset=[
        "Government Revenue (% of GDP)",
        "Government Expense (% of GDP)"
    ]
)

top_fiscal = year_filtered.sort_values(
    "Government Revenue (% of GDP)", ascending=False
).head(15)

fig = px.bar(
    top_fiscal,
    x="country_name",
    y=[
        "Government Revenue (% of GDP)",
        "Government Expense (% of GDP)"
    ],
    title="Government Revenue vs Expense (2020)",
    barmode="group"
)

fig.update_layout(
    template="plotly_white",
    title_x=0.5,
    xaxis_title="Country",
    yaxis_title="% of GDP"
)

fig.show()
```
